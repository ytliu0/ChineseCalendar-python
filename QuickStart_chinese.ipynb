## 公曆轉中曆

本軟體提供以下三個函數可用來轉公曆為中曆：

- `.western_to_chinese_date(y, m, d, ...)`: 公曆日期(y=年，m=月，d=日)轉中曆日期。

- `.western_to_chinese_year(y, m=None, ...)`: 公曆 y 年 m 月 中所有日期轉為中曆日期。如果 m 是 None，則將公曆年 y 中所有日期轉為中曆日期。

- `.western_to_chinese_year_html(y, m=None, ...)`: 與函數`.western_to_chinese_year(y, m=None, ...)`的功能一樣，但會輸出 HTML 格式的文字。

以上`...`表示函數還有其他可選的參數，但本指南不會討論，略去這些參數，函數執行時會取默認值。這些函數參數的用法在`tutorial/traditional_Chinese/tutorial.iynb`的深入指引中介紹。

以下指令將公曆2025年1月26日轉為中曆日期。

```python
cal.western_to_chinese_date(2025, 1, 26)
```

函數的輸出顯示公曆2025年1月26日是農曆2024年(甲辰年)十二月廿七日。

此函數的輸出是 python dictionary，有下列的字典鍵(keys):

`'Western ymd'`: python list 包含公曆的年月日。

`'Western calendar'`: 公曆版本 (格里曆、儒略曆或逆推儒略曆)。

`'JDN'`: 正午儒略日數。

`'Chinese ymd'`: 包含中曆的年月日，閏月的月份用負數表示。 

`'cyear'`: 與中曆年日期重疊最多的公曆年份，也是`'Chinese ymd'` list 中第一個數字。此例的甲辰年就是2024年，因為甲辰年始於2024年2月10日，終於2025年1月28日，甲辰年與公曆2024年有326日重疊，是重疊日數最多的公曆年，因此甲辰年的`'cyear'`取2024。

`'cdate'`: 中曆日期。

`'sexagenary year'`: python list 顯示年干支，年地支對應的生肖，及干支數 (1=甲子，2＝乙丑，3＝丙寅等，六十干支的干支數可參閱[干支網頁](https://ytliu0.github.io/ChineseCalendar/sexagenary_chinese.html)的表二) 。

`'jian'`: 中曆月的月建干支，月干支的詳情可參閱[干支網頁](https://ytliu0.github.io/ChineseCalendar/sexagenary_chinese.html)。

`'Nday_cmonth'`: 中曆月的日數，一般是29日或30日。

`'sexagenary date'`: 日干支及其干支數，日干支的詳情參閱[干支網頁](https://ytliu0.github.io/ChineseCalendar/sexagenary_chinese.html)。

`'Week'`: 星期及星期數，星期數用 0、1、2、3、4、5、6 表示星期日、一、二、三、四、五、六。 

`'period'`: 中國歷史的時代。

`'Chinese calendar'`: 中曆的版本。

與我的[農曆網站](https://ytliu0.github.io/ChineseCalendar/index_chinese.html)一樣，公曆年 y 的年限是 -721 至 2200。當中國處於分裂時代時常有多個曆法同時使用，這裡用的指令只提供其中一個版本，其他版本的日期也可以獲取，但要用上述`...`中的參數，本指南不會說明，如想知其用法，可參閱`tutorial/traditional_Chinese/tutorial.ipynb`的進階指引。

以下指令`cal.western_to_chinese_year(2025, 1)`輸出公曆2025年1月所有日期的中曆資料。指令 `cal.western_to_chinese_year(2025)`則顯示2025年全年的中曆日期資料。

```python
cal.western_to_chinese_year(2025, 1)
```

以上的輸出如果用於其後的日曆計算或許有用，但對於瀏覽來說並不是很方便，函數`.western_to_chinese_year_html()`輸出的 HTML 文字用`IPython.display.HTML()`的函數處理後，就會得到如[這農曆網頁](https://ytliu0.github.io/ChineseCalendar/index_chinese.html)的表列顯示，比較方便瀏覽。

```python
# 2025年11月的月曆表 
HTML(cal.western_to_chinese_year_html(2025, 11))
```

```python
# 1979年的年曆表
HTML(cal.western_to_chinese_year_html(1979))
```

```python
# 公曆 690 年的年曆表 
HTML(cal.western_to_chinese_year_html(690))
```

上面列出的月相和二十四節氣如[農曆網站](https://ytliu0.github.io/ChineseCalendar/index_chinese.html)的一樣是指用現代方法以噴射推進實驗室的 DE441 曆表算出，而曆書節氣指按當時中國曆法算出的節氣日期。

中國科學院紫金山天文臺從 2014年起用 DE421 曆表編算天文年曆，直至2025年還沒有更新曆表，不過 DE421 算出的月相和節氣時刻在現代相差甚微。所有用現代曆表算出的月相和節氣時刻都用東八區時間 UT+8 (UT 在1972年前指 UT1， 1972年到今年指 UTC，未來年份則是預推的 UTC)。

## 中曆轉公曆

本軟體提供以下四個函數可用來轉中曆為公曆：

`.chinese_to_western_date(y, m, d, ...)`: 轉中曆年月日 y, m, d 為公曆日期，閏月的 m 值為負數。

`.chinese_year_info(y, ...)`: 此函數輸出中曆年 y 的所有中曆月初一的公曆日期，曆書節氣，及用現代曆表 DE441 算出的節氣。

`.chinese_year_html(y, ...)`: 與上述函數的功能一致，但輸出 HTML 文字。

`.chinese_to_western_year(y, m=None, ...)`: 列出中曆年 y 中曆月 m 所有日期的公曆日期。如果 m 是 None，則列出中曆年全年的公曆日期。

`chinese_to_western_year_html(y, m=None, ...)`: 與上述函數的功能一致，但輸出 HTML 文字。

同上， `...`表示其他可選的參數，但本指南不會討論。

```python
# 轉 2025年農曆七月初七為公曆
cal.chinese_to_western_date(2025, 7, 7)
```

由於曆法變更，歷史上有少數情況某中曆日期對應兩個公曆日期。例如武周聖曆三年/久視元年(700年)的十一月(當時稱為正月)和十二月。以下指令顯示700年十一月廿三日對應公曆日期是699年12月19日及701年1月6日。

```python
cal.chinese_to_western_date(700, 11, 23)
```

以下指令輸出中曆700年的曆日資料。

```python
cal.chinese_year_info(700)
```

此函數輸出 python dictionary，有以下字鍵：

`'cm'`: python list 含有以下中曆資料： \[{'cm':m1, 'JDN':JDN1, 'jian':'jian1', 'n':n1}, {'cm':m2, 'JDN':JDN2, 'jian':'jian2', 'n':n2}, ...\]，'cm' 表示中曆月序(負數表示閏月), 'JDN' 表示朔日的正午儒略日數， 'jian' 表示用月建來算的月干支數 ('jian'='1' 表示甲子，'2' 表示乙丑，⋯⋯，'60' 表示癸亥，'NA' 表示中曆月沒有月建，一般是閏月，但-103年以前的曆法因不用無中氣置閏也不列月建，'NZ' 表示此月不含中氣，這只適用於古六曆的月份)，'n' 表示中曆月的日數。 此 python list 以 'JDN' 的數值由小至大排序。

`'calendar'`: 中曆的版本 (例如 '三統曆'， '時憲曆' 等)。

`'reign/era'` (只適用於 1912年前的年份): 中曆年的年號或皇帝紀年。

`'calst'`: 曆書節氣 tuples \[(lab1, jdn1), (lab2, jdn2), ...\]， lab1, lab2, ... 是曆書節氣的標記，例如 'J1' 表示正月節(立春)，'Z4' 表示四月中(小滿) 等，jdn1, jdn2, ... 是曆書節氣的正午儒略日數。

`'datongST'` (只適用於1667至1669的中曆年): 用大統曆算出的節氣 \[(lab1, jdn1), (lab2, jdn2), ...\]。康熙六年(1667)至八年(1669)清朝因曆獄而復用明朝《大統曆》，但後來清朝欽天監編的《御定萬年書》試圖抹去這段歷史，用時憲曆追推這幾年的節氣，但臺灣國立故宮博物院的清代檔案檢索系統藏有這幾年的《大清七政經緯躔度時憲曆》，顯示這幾年的節氣日期確是符合大統曆的推算，而不符合時憲曆的推算，所以此處在這幾年列出大統曆的節氣。

`'sterm'`: python list \[\[lab1, info1\], \[lab2, info2\], ...\] 顯示用 DE441曆表算出的節氣，lab1, lab2, ... 為節氣標記，info1, info2, ... 則是 python dictionary，顯示節氣的日期及 UT+8 時刻。

`'mphase'`: python list \[\[lab1, info1\], \[lab2, info2\], ...\] 顯示用 DE441 曆表算出的中曆年內所有的月相資料。lab1, lab2, ... 是月相的標記： 'Q0' 是朔，'Q1' 是上弦，'Q2' 是望，'Q3' 是下弦。 info1, info2, ... 是 python dictionary，顯示月相的日期及 UT+8 時刻。

此函數輸出的 python list 不方便瀏覽， 函數 `.chinese_year_info_html()` 輸出 HTML 文字，可用 `IPython.display.HTML()` 函數將資料以表格形式顯示。

```python
# 700年的中曆年資料
HTML(cal.chinese_year_html(700))
```

```python
# 1669年的中曆年資料
HTML(cal.chinese_year_html(1669))
```

```python
# 2025年的中曆年資料
HTML(cal.chinese_year_html(2025))
```

```python
# 中曆156年九月的所有日期
cal.chinese_to_western_year(156, 9)
```

```python
# 中曆156年九月的月曆表 
HTML(cal.chinese_to_western_year_html(156, 9))
```

```python
# 中曆1980年的年曆表 
HTML(cal.chinese_to_western_year_html(1980))
```

注意此處列出的月相和節氣日期是中曆日期，而不是公曆日期。

## 月相和二十四節氣

`ChineseCalendar_py.sunMoon.sunMoon` class 提供用現代曆表算出的月相和二十四節氣的 UT+8 日期和時刻，以下指令建立這 class 的實體。

```python
sm = ccal.sunMoon.sunMoon('ChiT')
```

第一個輸入參數 'ChiT' 表示輸出的語言是傳統中文，其他可選的是 'Eng'(英文)和 'ChiS' (簡體中文)。

第二個可輸入的參數是 'DE431' 或 'DE441'， 表示計算月相和節氣用的曆表，默認值是 'DE441'。如前述，紫金山天文臺自2014年起用 DE421 曆表編算天文年曆，直至2025年仍是如此，DE421、DE431及DE441曆表算出的月相和節氣在距今幾十年間差異甚微(DE421 曆表的年限是1899年至2053年)，在遠古和遙遠將來 DE431 和 DE441 算出的時刻雖然差異較大，但仍未超出 TDB 轉換 UT1 的估計誤差。1929年至2013年中曆的月相和節氣用舊曆表計算，與 DE 曆表的差異可達半分鐘。

```python
# 2025年全年的月相和節氣
st, mp = sm.st_mp(2025)
print('節氣:\n', st)
print('月相:\n', mp)
```

```python
# 2025年全年的月相和節氣，轉為 pandas dataframe 來顯示。
st, mp = sm.st_mp_df(2025)
display(st)
display(mp)
```

月相中朔和望還提供日月食的資料，有連結到我製的日月食網頁，要恰當顯示連結，可用 pandas dataframe 提供的 `.to_html` 函數轉為 HTML 文字，然後用 `IPython.display.HTML()` 顯示，得出的輸出類此[氣朔網頁](https://ytliu0.github.io/ChineseCalendar/sunMoon_chinese.html)的表格。

```python
HTML(mp.to_html(
    render_links=True,
    escape=False,
))
```

## 曆書節氣

古代的節氣用當時的曆法計算，算出的節氣與用現代曆表的節氣常有大差異，函數`.calendrical_solar_terms(y)`提供公曆 y 年的曆書節氣，即按當時曆法計算的節氣。

```python
# 1532年的曆書節氣
cal.calendrical_solar_terms(1532)
```

```python
# 1532年的曆書節氣轉為 pandas dataframe 顯示 
display(pd.DataFrame(cal.calendrical_solar_terms(1532)))
```

```python
# -106年(公元前107年)的曆書節氣
display(pd.DataFrame(cal.calendrical_solar_terms(-106)))
```

此函數輸出公曆 y 年的曆書節氣，如要中曆 y 年的曆書節氣，可用上述
`.chinese_year_info(y)` 或 `.chinese_year_html(y)` 函數。
