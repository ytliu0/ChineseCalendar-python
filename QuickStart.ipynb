## Western calendar to Chinese calendar conversion

Three methods are provides to do the conversion:

- `.western_to_chinese_date(y, m, d, ...)`: Western date (year = y, month = m, date = d) to Chinese date.

- `.western_to_chinese_year(y, m=None, ...)`: Western year y, month m (if m is not None) to Chinese dates

- `.western_to_chinese_year_html(y, m=None, ...)`: Same as `western_to_chinese_year()` but return an HTML text.

Note that `...` denotes other optional parameters not covered in this guide. See the tutorial notebook for detail.

The following command converts January 26, 2025 to Chinese date.

```python
cal.western_to_chinese_date(2025, 1, 26)
```

This shows that January 26, 2025 corresponds to the 27th day in the 12th month in the Chinese year Jia Chen (the year of Dragon). 

This method `.western_to_chinese_date(y, m, d)` returns a python dictionary with the following keys:

`'Western ymd'`: a list containing Western year, month, day

`'Western calendar'`: version of the Western calendar (Gregorian, Julian or propleptic Julian).

`'JDN'`: Julian date at noon. 

`'Chinese ymd'`: a list containing Chinese year, month (negative means leap month) and date. 

`'cyear'`: Western year with the most overlap of days in the Chinese year.

`'cdate'`: Chinese month number-day number. 

`'sexagenary year'`: a list with \[sexagenary year cycle, sexagenary number\]. Sexagenary number 1 means Jia zi, 2 means Yi chow, and so on (see Table 2 on [this page](https://ytliu0.github.io/ChineseCalendar/sexagenary.html) for the complete list). 

`'jian'`: sexagenary month cycle. See [this page](https://ytliu0.github.io/ChineseCalendar/sexagenary.html) for an explanation of the sexagenary month cycle and *jian*.

`'Nday_cmonth'`: number of days in the Chinese month, usually 29 or 30.

`'sexagenary date'`: sexagenary date cycle. See [this page](https://ytliu0.github.io/ChineseCalendar/sexagenary.html) for an explanation of the sexagenary date cycle.

`'Week'`: \[day of week, week number\]. Week number convention: 0 for Sunday, 1 for Monday, 2 for Tuesday, 3 for Wednesday, 4 for Thursday, 5 for Friday and 6 for Saturday.

`'period'`: Period in the Chinese history.

`'Chinese calendar'`: version of the Chinese calendar.

As in [my Chinese calendar webpage](https://ytliu0.github.io/ChineseCalendar), the year y should be between -721 and 2200. In some years in the past, there were multiple versions of Chinese calendar used in those years, and the command returns only one of them. Multiple versions of calendar dates can be retrieved using parameters `...` not covered here. See the tutroial notebook for detail. 

The following command `cal.western_to_chinese_year(2025, 1)` returns a list of dictaionary showing all dates in January 2025. The command `cal.western_to_chinese_year(2025)` shows all dates in 2025.

```python
cal.western_to_chinese_year(2025, 1)
```

The list above may be useful for further calendar calculations but is not convenient to view. The method `.western_to_chinese_year_html()` returns an html text, which can be displayed nicely in a Jupyter notebook using the `IPython.display.HTML()` function. The result is similar to the calendar table on [this Chinese calendar webpage](https://ytliu0.github.io/ChineseCalendar).

```python
# Calendar table for November, 2025
HTML(cal.western_to_chinese_year_html(2025, 11))
```

```python
# Calendar table for the year 1979
HTML(cal.western_to_chinese_year_html(1979))
```

```python
# Calendar table for the year 690
HTML(cal.western_to_chinese_year_html(690))
```

As in [my Chinese calendar page](https://ytliu0.github.io/ChineseCalendar), moon phases and solar terms refer to the modern calculation using JPL's DE441 ephemeris, whereas calendrical solar terms refer to the solar terms calculated using the method at that time. Starting from 2014 and as of 2025, the official moon phases and solar terms promulgated by the Purple Mountain Observatory (the calendrical moon phases and solar terms of modern times) are computed using JPL's DE421/LE421, which are very close to our results based on DE441. All times calculated by DE441 are in UT+8 (UT1+8 for years before 1972, UTC+8 for years between 1972 and present year, and projected UTC+8 for future years).

## Chinese calendar to Western calendar conversion

Four methods are provided to do the conversion:

`.chinese_to_western_date(y, m, d, ...)`: Given a Chinese date (year = y, month = m, date = d), convert to the date in the Western calendar. For a leap month, enter a negative number for `m`.

`.chinese_year_info(y, ...)`: Given a Chinese year whose New Year day is closest to Jan 1, y in the Western calendar, this function returns a list containing the first days in Chinese months, calendrical solar terms and solar terms computed by a DE ephemeris in the Chinese year.

`.chinese_year_html(y, ...)`: Same as the previous function but returns an HTML text.

`.chinese_to_western_year(y, m=None, ...)`: Given a Chinese year y, return a list containing the Western dates corresponding to all days in the year if m is None or Western dates corresponding to all days in month m in the Western calendar.

`chinese_to_western_year_html(y, m=None, ...)`: Same as the previous function but returns an HTML text.

Again `...` denotes other optional parameters not covered in this guide.

```python
# Convert the 7th day in the 7th month in Chinese year 2025 to Western date
cal.chinese_to_western_date(2025, 7, 7)
```

Ocassionally there are multiple days corresponding to the same Chinese date in the past. This is caused by a Chinese calendar reform. As an example, let's look at the 23rd day in the 11th month in the Chinese year 700.

```python
cal.chinese_to_western_date(700, 11, 23)
```

The following command returns the information of the Chinese year 700.

```python
cal.chinese_year_info(700)
```

The function returns a python dictionary with the following keys:

`'cm'`: a list containing the information of all Chinese months in the form of \[{'cm':m1, 'JDN':JDN1, 'jian':'jian1', 'n':n1}, {'cm':m2, 'JDN':JDN2, 'jian':'jian2', 'n':n2}, ...\] with 'cm' indicating the Chinese month number (negative for a leap month), 'JDN' indicating the Julian date at noon of the first day of the Chinese month, 'jian' indicating the sexagenary month cycle associated with jian ('jian'='1' for jia zi, '2' for yi chou, ... '60' for gui hai, 'NA' for months without jian (leap months or months before -103 when no Zhongqi rule was not imposed), 'NZ' for a month not containing a major solar term (only for the ancient six calendars)), and 'n' indicating the number of days in the Chinese month. The list is sorted in ascending order in the values in 'JDN'.

`'calendar'`: version of the Chinese calendar (e.g. 'Santong', 'Shixian').

`'reign/era'` (optional): reign and/or era name of the Chinese year (only for years before 1912).

`'calst'`: list of tuples \[(lab1, jdn1), (lab2, jdn2), ...\] where lab1, lab2, ... are the lables of the calendrical solar terms (e.g. 'J1', 'Z4') and jdn1, jdn2, ... are the Julian dates at noon of the solar terms.

`'datongST'` (optional): list of tuples \[(lab1, jdn1), (lab2, jdn2), ...\] of all the Datong solar terms in the Chinese year. This is only present for 1667 <= y <= 1669.

`'sterm'`: a list \[\[lab1, info1\], \[lab2, info2\], ...\] containing all solar terms based on DE441 in Chinese year y, where lab1, lab2, ... are the labels of the solar terms (e.g. 'J8', 'Z3') and info1, info2, ... are dictionaries containing the Western date and time information of the solar terms.

`'mphase'`: a list \[\[lab1, info1\], \[lab2, info2\], ...\] containing all moon phases based on DE441 in year y, where lab1, lab2, ... are the labels of the moon phases (e.g. 'Q0', 'Q3') and info1, info2, ... are dictionaries containing the Western date and time information of the moon phases.

The returned python list is not convenient for viewing. The `.chinese_year_info_html()` method geneartes an HTML text for a table, which can be viewed in the Jupyter notebook using `IPython.display.HTML()` function.

```python
# Chinese year 700
HTML(cal.chinese_year_html(700))
```

```python
# Chinese year 1669
HTML(cal.chinese_year_html(1669))
```

```python
# Chinese year 2025
HTML(cal.chinese_year_html(2025))
```

```python
# Extract all days in Chinese month 9 in the Chinese year 156
cal.chinese_to_western_year(156, 9)
```

```python
# Create an HTML table for the Chinese month 9 in the Chinese year 156
HTML(cal.chinese_to_western_year_html(156, 9))
```

```python
# Create an HTML table for the Chinese year 1980
HTML(cal.chinese_to_western_year_html(1980))
```

Unlike `.western_to_chinese_year_html()`, the dates of moon phases and solar terms under each month refer to the dates in the Chinese calendar.

## Moon phases and solar terms

Dates and times of moon phases and solar terms calculated by a modern ephemeris can be obtained from the `.sunMoon.sunMoon` class. The following commands create an instance of the `sunMoon` object.

```python
sm = ccal.sunMoon.sunMoon('Eng', 'DE441')
```

The first parameter 'Eng' sets the display language as English. Two other possible parameters are 'ChiT' for traditional Chinese and 'ChiS' for simplified Chinese.

The second parameter 'DE441' sets the ephemeris for moon phases and solar terms to be JPL's DE441 ephemeris. Another option is 'DE431' for JPL's DE431 ephemeris. These two ephemerides give practically identical moon phase and solar term times in years within a few hundred years from 2000. As mentioned above, the official moon phases and solar terms promulgated by the Purple Mountain Observatory in 2014-present (as of 2025) are calculated by JPL's DE421/LE421, which also gives practially identical moon phase and solar term times from 2014-present. The moon phases and solar terms promulgated by the Purple Mountain Observatory in 1951-2013 were calculated by older ephemerides and could deviate from those of DE431/DE441 by one minute.

```python
# Extract solar terms and moon phases in 2025
st, mp = sm.st_mp(2025)
print('solar terms:\n', st)
print('moon phases:\n', mp)
```

```python
# Extract solar terms and moon phases in 2025 and put them into pandas df's
st, mp = sm.st_mp_df(2025)
display(st)
display(mp)
```

Note that links to information on eclipses are provided in the moon phase data. To properly display the links, we can convert the pandas df to html and then use HTML to display the result. The resulting table is similar to my [moon phase and solar term page](https://ytliu0.github.io/ChineseCalendar/sunMoon.html).

```python
HTML(mp.to_html(
    render_links=True,
    escape=False,
))
```

## Calendrical solar terms

Solar terms in the past were not calculated using modern ephemerides, but were calculated using the astronomical systems developed in the past. I call them the calendrical solar terms. The method `.calendrical_solar_terms(y)` returns the calendrical solar terms in Western year y.

```python
# calendrical solar term in 1532
cal.calendrical_solar_terms(1532)
```

```python
# calendrical solar term in 1532 displayed as a pandas df
display(pd.DataFrame(cal.calendrical_solar_terms(1532)))
```

```python
# calendrical solar terms for -106 (107 BCE)
display(pd.DataFrame(cal.calendrical_solar_terms(-106)))
```

Calendrical solar terms in a Chinese year can be obtained using the method `.chinese_year_info(y)` or `.chinese_year_html(y)` shown above.
