## Default and Available Calendars

Over the course of Chinese history, there were periods where multiple Chinese calendars were used in different regions in China. This package supports multiple, but not all, versions of Chinese calendars in those periods. 

To prevent confusion, one set of Chinese calendars is selected in those periods used by particular states/regions. This is called the *default calendars* here. In the past, those states were regarded as the legitimate governments representing China by some Chinese historians. Today they are regarded as just conventions with no legitimacy connotation. The [defaule calendar page](https://ytliu0.github.io/ChineseCalendar-python/tutorial/English/default_calendar.html) lists the versions of Chinese calendars selected by the default calendar, which covers the years from -721 (722 BCE) to 2200.

As stated on the [defaule calendar page](https://ytliu0.github.io/ChineseCalendar-python/tutorial/English/default_calendar.html), the calendar data in -103 – 2200 are stored in an array for table lookup. The data are stored in the source-code folder in the file 'default_calendar_M105_2201.csv', which contains the conjunction dates, leap month information and calendrical solar terms counted from Jan 0 of the year. When an instance of the Chinese calendar class is created, the conjunction and leap month information are loaded to the `.calData` array and the calendrical solar term data are loaded to the `calSterm` array. These arrays may be useful in some calendar calculations, as shown in the calendar exercise notebook in this folder. The year span in the arrays are actually from -105 to 2202 for the convenience of calendar calculations.

Apart from the calendars used by the default calendars, this package also supports a number of other version of Chinese calendars. The [available calendar page](https://ytliu0.github.io/ChineseCalendar-python/tutorial/English/available_calendars.html) lists all supported calendars. Note that all the calendars there were used in certain period and certain regions in China. Most of them are calculated based on their calendar rules and can in principle be used outside the years of usage, but it is strongly discouraged to do so.

## Western calendar to Chinese calendar conversion

Three methods are provides to do the conversion:

- `.western_to_chinese_date(y, m, d, calendar=default, multiple_calendars=False, ephemeris='None')`: Western date (year = y, month = m, date = d) to Chinese date.

- `.western_to_chinese_year(y, m=None, calendar='default', ephemeris='None')`: Western year y, month m (if m is not None) to Chinese dates

- `.western_to_chinese_year_html(y, m=None, calendar='default', showJDN=False, ephemeris='None', show_chinese_year_info=True,show_chinese_calendar_info=True)`: Same as `western_to_chinese_year()` but return an HTML text.

The 'multiple_calendars' argument in `.western_to_chinese_date()` is only for the default calendars. If set to True, it will return the dates from multiple Chinese calendars in the periods when China was divided.

The 'calendar' argument allows the user to select a particular Chinese calendar. It's either 'default' (for the default calendars) or one of the calendar labels listed on the [available calendar page](https://ytliu0.github.io/ChineseCalendar-python/tutorial/English/available_calendars.html).

The 'ephemeris' argument allows a user to specify an ephemeris for the times of modern moon phases and solar terms. It's either DE441 or DE431. When the argument is missing, it will take the default value 'None', which will pick whatever ephemeris previously specified or DE441 if an ephemeris has not been specified.

The following command converts February 25, 445 to the Chinese calendar used in the Northern Wei dynasty. According to the [available calendar page](https://ytliu0.github.io/ChineseCalendar-python/tutorial/English/available_calendars.html), we need to use the 'WeiZhouSui' label.

```python
cal.western_to_chinese_date(445, 2, 25, calendar='WeiZhouSui')
```

This shows that February 25, 445 corresponds to the 3rd day in the leap month 1 in the Northern Wei calendar, which was based on the Jingchu system. The Chinese year was Tàiwǔdì Tàipíngzhēnjūn 6 and its sexagenary year cycle was Yi You.

This method `.western_to_chinese_date()` returns a python dictionary with the following keys:

`'Western ymd'`: a list containing Western year, month, day

`'Western calendar'`: version of the Western calendar (Gregorian, Julian or propleptic Julian).

`'JDN'`: Julian date at noon. 

`'Chinese ymd'`: a list containing Chinese year, month (negative means leap month) and date. 

`'cyear'`: Western year with the most overlap of days in the Chinese year.

`'cdate'`: Chinese month number-day number. 

`'sexagenary year'`: a list with \[sexagenary year cycle, sexagenary number\]. Sexagenary number 1 means Jia zi, 2 means Yi chow, and so on (see Table 2 on [my sexagenary cycle page](https://ytliu0.github.io/ChineseCalendar/sexagenary.html) for the complete list). 

`'jian'`: sexagenary month cycle. See [my sexagenary cycle page](https://ytliu0.github.io/ChineseCalendar/sexagenary.html) for an explanation of the sexagenary month cycle and *jian*.

`'Nday_cmonth'`: number of days in the Chinese month, usually 29 or 30.

`'sexagenary date'`: sexagenary date cycle. See [my sexagenary cycle page](https://ytliu0.github.io/ChineseCalendar/sexagenary.html) for an explanation of the sexagenary date cycle.

`'Week'`: \[day of week, week number\]. Week number convention: 0 for Sunday, 1 for Monday, 2 for Tuesday, 3 for Wednesday, 4 for Thursday, 5 for Friday and 6 for Saturday.

`'period'`: Period in the Chinese history.

`'Chinese calendar'`: version of the Chinese calendar.

As in [my Chinese calendar webpage](https://ytliu0.github.io/ChineseCalendar), the year y should be between -721 and 2200. In some years in the past, there were multiple versions of Chinese calendar used in those years, and the command returns only one of them. Multiple versions of calendar dates can be retrieved by setting calendar='default' and multiple_calendars=True:

```python
cal.western_to_chinese_date(445, 2, 25, multiple_calendars=True)
```

This output shows that February 25, 445 also corresponds to the 3rd day in the second month in the calendar used by the Song dynasty, which was based on the Yuanjia system. This is the date that is returned for calendar='default'.

The following command `cal.western_to_chinese_year(445, 2, calendar='WeiZhouSui')` returns a list of dictaionary showing all dates in February 445 in the Northern Wei calendar. The command `cal.western_to_chinese_year(445, calendar='WeiZhouSui')` shows all dates in 445.

```python
cal.western_to_chinese_year(445, 2, calendar='WeiZhouSui')
```

The list above may be useful for further calendar calculations but is not convenient to view. The method `.western_to_chinese_year_html()` returns an html text, which can be displayed nicely in a Jupyter notebook using the `IPython.display.HTML()` function. The result is similar to the calendar table on [my Chinese calendar webpage](https://ytliu0.github.io/ChineseCalendar).

```python
# Calendar table for February, 445 in the Northern Wei calendar
HTML(cal.western_to_chinese_year_html(445, 2, calendar='WeiZhouSui'))
```

```python
# Showing the JDN (Julian date at noon) in the calendar table February, 445 
# in the Northern Wei calendar
HTML(cal.western_to_chinese_year_html(445, 2, calendar='WeiZhouSui', showJDN=True))
```

As in [my Chinese calendar page](https://ytliu0.github.io/ChineseCalendar), moon phases and solar terms refer to the modern calculation using JPL's DE441 ephemeris, whereas calendrical solar terms refer to the solar terms calculated using the method at that time. Starting from 2014 and as of 2025, the official moon phases and solar terms promulgated by the Purple Mountain Observatory (the calendrical moon phases and solar terms of modern times) are computed using JPL's DE421/LE421, which are very close to our results based on DE441. All times calculated by DE431/DE441 are in UT+8 (UT1+8 for years before 1972, UTC+8 for years between 1972 and present year, and projected UTC+8 for future years).

```python
# Show the calenar table for 2500 computed by GB/T 33661-2017.
HTML(cal.western_to_chinese_year_html(2500, calendar='GBT'))
```

## Chinese calendar to Western calendar conversion

Four methods are provided to do the conversion:

`.chinese_to_western_date(y, m, d, calendar='default', ephemeris='None')`: Given a Chinese date (year = y, month = m, date = d), convert to the date in the Western calendar. For a leap month, enter a negative number for `m`.

`.chinese_year_info(y, calendar='default', ephemeris='None')`: Given a Chinese year whose New Year day is closest to Jan 1, y in the Western calendar, this function returns a list containing the first days in Chinese months, calendrical solar terms and solar terms computed by a DE ephemeris in the Chinese year.

`.chinese_year_html(y, calendar='default', ephemeris='None')`: Same as the previous function but returns an HTML text.

`.chinese_to_western_year(y, m=None, calendar='default', ephemeris='None')`: Given a Chinese year y, return a list containing the Western dates corresponding to all days in the year if m is None or Western dates corresponding to all days in month m in the Western calendar.

`chinese_to_western_year_html(y, m=None, calendar='default', showJDN=False, ephemeris='None', show_chinese_year_info=True, show_chinese_calendar_info=True)`: Same as the previous function but returns an HTML text.

There are no equivalent pages in [my Chinese calendar website](https://ytliu0.github.io/ChineseCalendar) providing the same information as these functions. The closest one is the [conversion table page](https://ytliu0.github.io/ChineseCalendar/table.html), which provides information similar to the function `.chinese_year_html()`. 

The usage of these functions are similar to the functions for the conversion from Western to Chinese calendar. Below I show a few examples.

```python
# Convert the 15th day in the 8th month in Chinese year 1661 in the Southern Ming dynasty 
# to Western date
cal.chinese_to_western_date(1661, 8, 15, calendar='SouthernMing')
```

This shows that the Western date was September 8, 1661. Note that this was not the date in the calendar used by the Qing dynasty, which was October 7, 1661 as shown by the command `cal.chinese_to_western_date(1661, 8, 15)`.

Ocassionally there are multiple days corresponding to the same Chinese date in the past. This is caused by a Chinese calendar reform. As an example, let's look at the 12th day in the 5th month in the Chinese year 762.

```python
cal.chinese_to_western_date(762, 5, 12)
```

The following command returns the information of the Chinese year -250 (251 BCE) in the Zhuanxu calendar.

```python
cal.chinese_year_info(-250, calendar='Zhuanxu')
```

The function returns a python list with the following keys:

`'cm'`: a list containing the information of all Chinese months in the form of \[{'cm':m1, 'JDN':JDN1, 'jian':'jian1', 'n':n1}, {'cm':m2, 'JDN':JDN2, 'jian':'jian2', 'n':n2}, ...\] with 'cm' indicating the Chinese month number (negative for a leap month), 'JDN' indicating the Julian date at noon of the first day of the Chinese month, 'jian' indicating the sexagenary month cycle associated with jian ('jian'='1' for jia zi, '2' for yi chou, ... '60' for gui hai, 'NA' for months without jian (leap months or months before -103 when no Zhongqi rule was not imposed), 'NZ' for a month not containing a major solar term (only for the ancient six calendars)), and 'n' indicating the number of days in the Chinese month. The list is sorted in ascending order in the values in 'JDN'.

`'calendar'`: version of the Chinese calendar (e.g. 'Santong', 'Shixian').

`'reign/era'` (optional): reign and/or era name of the Chinese year (only for years before 1912).

`'calst'`: list of tuples \[(lab1, jdn1), (lab2, jdn2), ...\] where lab1, lab2, ... are the lables of the calendrical solar terms (e.g. 'J1', 'Z4') and jdn1, jdn2, ... are the Julian dates at noon of the solar terms.

`'datongST'` (optional): list of tuples \[(lab1, jdn1), (lab2, jdn2), ...\] of all the Datong solar terms in the Chinese year. This is only present for 1667 <= y <= 1669.

`'sterm'`: a list \[\[lab1, info1\], \[lab2, info2\], ...\] containing all solar terms based on DE431/DE441 in Chinese year y, where lab1, lab2, ... are the labels of the solar terms (e.g. 'J8', 'Z3') and info1, info2, ... are dictionaries containing the Western date and time information of the solar terms.

`'mphase'`: a list \[\[lab1, info1\], \[lab2, info2\], ...\] containing all moon phases based on DE431/DE441 in year y, where lab1, lab2, ... are the labels of the moon phases (e.g. 'Q0', 'Q3') and info1, info2, ... are dictionaries containing the Western date and time information of the moon phases.

The returned python list is not convenient for viewing. The `.chinese_year_info_html()` method geneartes an HTML text for a table, which can be viewed in the Jupyter notebook using `IPython.display.HTML()` function.

```python
# Chinese year -250 in the Zhuanxu calendar
HTML(cal.chinese_year_html(-250, calendar='Zhuanxu'))
```

```python
# Chinese year 2714 based on GB/T 33661-2017
HTML(cal.chinese_year_html(2714, calendar='GBT'))
```

```python
# Extract all days in Chinese month 8 in the Chinese year 955 in the calendar used by the Liao dynasty
cal.chinese_to_western_year(955, 8, calendar='LiaoJinYuan')
```

```python
# Create an HTML table for the Chinese month 8 in the Chinese year 955 
# in the calendar used by the Liao dynasty.
HTML(cal.chinese_to_western_year_html(955, 8, calendar='LiaoJinYuan'))
```

```python
# Create an HTML table for the Chinese year 254
# in the calendar used by the Shu state.
HTML(cal.chinese_to_western_year_html(254, calendar='Shu'))
```

Unlike `.western_to_chinese_year_html()`, the dates of moon phases and solar terms under each month refer to the dates in the Chinese calendar.

## Moon phases and solar terms

Dates and times of moon phases and solar terms calculated by a modern ephemeris can be obtained from the `ChineseCalendar_py.sunMoon.sunMoon` class. The following commands create an instance of the `sunMoon` object.

```python
sm = ccal.sunMoon.sunMoon('Eng')
```

The first parameter 'Eng' sets the display language as English. Two other possible parameters are 'ChiT' for traditional Chinese and 'ChiS' for simplified Chinese.

There is an optional second parameter (either 'DE431' or 'DE441') that allows the user to choose an ephemeris. When the second parameter is absent, it takes on the default value 'DE441' for JPL's DE441 ephemeris. for moon phases and solar terms to be JPL's DE441 ephemeris. DE441 and DE431 give practically identical moon phase and solar term times in years within a few hundred years from 2000. As mentioned above, the official moon phases and solar terms promulgated by the Purple Mountain Observatory in 2014-present (as of 2025) are calculated by JPL's DE421/LE421, which also gives practially identical moon phase and solar term times from 2014-present. The moon phases and solar terms promulgated by the government in 1929-2013 were calculated by older ephemerides and could deviate from those of DE431/DE441 by half minutes. Before 1929, moon phases and solar terms not only were calculated by old ephemerides but times were not given in UT+8. 

The TDB times of moon phases and solar terms in -3502-3503 are pre-computed by DE431 and DE441. They are stored in the files 'sunMoon_DE431.csv' and 'sunMoon_DE441.csv' in the folder `src/ChineseCalendar_py/`. Times in UT+8 are converted from the TDB times in the function `get_dTT_UT()` in `src/ChineseCalendar_py/` folder. This function will be updated on a yearly basis to incorporate the most up-to-date leap second data. As leap seconds will be abolished before 2035, the updates will be much less frequent after 2035.

```python
# Extract solar terms and moon phases in 3500 
st, mp = sm.st_mp(3500)
print('solar terms:\n', st)
print('moon phases:\n', mp)
```

```python
# Extract solar terms and moon phases in 3500 and put them into pandas df's
st, mp = sm.st_mp_df(3500)
display(st)
display(mp)
```

Note that links to information on eclipses are provided in the moon phase data. To properly display the links, we can convert the pandas df to html and then use HTML to display the result. The resulting table is similar to my [moon phase and solar term page](https://ytliu0.github.io/ChineseCalendar/sunMoon.html).

```python
HTML(mp.to_html(
    render_links=True,
    escape=False,
))
```

## Calendrical solar terms

Solar terms in the past were not calculated using modern ephemerides, but were calculated using the astronomical systems developed in the past. I call them the calendrical solar terms. The method `.calendrical_solar_terms(y, calendar="default", ephemeris='None')` returns the calendrical solar terms in Western year y. The argument 'ephemeris' is only relevant for the GBT calendar.

```python
# calendrical solar term in 250 in the Wu state
cal.calendrical_solar_terms(250, calendar='Wu')
```

```python
# calendrical solar term in 250 in the Wu state displayed as a pandas df
display(pd.DataFrame(cal.calendrical_solar_terms(250, calendar='Wu')))
```

Calendrical solar terms in a Chinese year can be obtained using the method `.chinese_year_info()` or `.chinese_year_html()` shown above.
