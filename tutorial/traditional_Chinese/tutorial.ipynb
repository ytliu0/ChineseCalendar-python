## 預設中曆及其他支持的中曆

中國有時處於分裂局面，各政權行用不同的曆法，本軟體在分裂時代提供多部（但非全部）同時行用的曆法。

為免混亂，如果用者沒有指明用某一特定曆法，本軟體在中國分裂時代選取了一套曆法，稱為「預設中曆」。預設中曆在春秋時代用魯國曆法（春秋曆），戰國時代用周朝曆法，三國時代用魏國曆法，東晉南北朝用南朝（東晉、宋、齊、梁、陳）曆法，五代十國用北朝（梁、唐、晉、漢、周）曆法，宋、遼、金、蒙古時期用宋朝曆法，南明、明鄭、清時期用清朝曆法，詳情可參閱[預設中曆網頁](https://ytliu0.github.io/ChineseCalendar-python/tutorial/traditional_Chinese/default_calendar.html)，預設中曆的年限是-721年至2200年。

如[預設中曆網頁](https://ytliu0.github.io/ChineseCalendar-python/tutorial/traditional_Chinese/default_calendar.html)所述，-103年至2200年的中曆用查表法獲取，這段時期中曆的朔閏及曆書節氣數據存於源碼文件夾'src/ChineseCalendar_py/'的'default_calendar_M105_2201.csv'檔案中，曆日以距公曆年1月0日的日數表示。當中曆物件實例建立時，檔案中的朔閏數據載入 numpy array `.calData`，曆書節氣數據載入 numpy array `.calSterm`。在這資料夾的日曆計算習題 notebook 中會有實例展示這兩個 numpy array 在某些計算上很有用。為了方便計算日曆，這兩個 numpy array 的曆日資料其實涵蓋-105年至2202年。

除了預設中曆外，本軟體還提供一些其他的中曆，詳情可參閱[其他中曆網頁](https://ytliu0.github.io/ChineseCalendar-python/tutorial/traditional_Chinese/available_calendars.html)。網頁中列出的所有中曆都有行用年份，大多數中曆用其曆法法則計算，原則上可用於行用年份之外，但這裡極不鼓勵用者如此做。

## 公曆轉中曆

本軟體套件提供下列三個函數轉公曆為中曆：

- `.western_to_chinese_date(y, m, d, calendar='default', multiple_calendars=False, ephemeris='None')`: 公曆日期(年 = y, 月 = m, 日 = d)轉中曆日期。

- `.western_to_chinese_year(y, m=None, calendar='default', ephemeris='None')`: 公曆 y 年 m 月 中所有日期轉為中曆日期。如果 m 是 None，則將公曆年 y 中所有日期轉為中曆日期。

- `.western_to_chinese_year_html(y, m=None, calendar='default', showJDN=False, ephemeris='None', show_chinese_year_info=True,show_chinese_calendar_info=True)`: 與函數`.western_to_chinese_year(y, m=None, calendar='default', ephemeris='None')`的功能一樣，但會輸出 HTML 格式的文字。

`calendar`參數用來選取某中曆，默認值是'default'，表示預設中曆，用者可選擇[其他中曆網頁](https://ytliu0.github.io/ChineseCalendar-python/tutorial/traditional_Chinese/available_calendars.html)列出的中曆標記選擇其他中曆。

`multiple_calendars`參數只有當`calendar`參數是'default'(預設中曆)時才有作用。如果`multiple_calendars`設定為`True`，分裂時代的輸出中曆會列出多個當時行用的中曆。

`ephemeris`參數讓用者選擇用哪個現代曆表計算月相和節氣時刻，這裡只支持'DE431'和'DE441'，默認值是'None'，表示會用已設定的曆表，如果還未設定曆表，則用默認值'DE441'。

以下指令轉公曆445年2月25日為北魏行用的曆法。根據[其他中曆網頁](https://ytliu0.github.io/ChineseCalendar-python/tutorial/traditional_Chinese/available_calendars.html)，北魏的日曆要用 'WeiZhouSui' 標記。

```python
cal.western_to_chinese_date(445, 2, 25, calendar='WeiZhouSui')
```

函數輸出顯示公曆445年2月25日，在北魏的日曆是太武帝太平真君六年閏正月初三，年干支是乙酉、日干支是癸亥，當時北魏行用的曆法是景初曆。

此函數的輸出是 python dictionary，有下列的字典鍵(keys):

`'Western ymd'`: python list 包含公曆的年月日。

`'Western calendar'`: 公曆版本 (格里曆、儒略曆或逆推儒略曆)。

`'JDN'`: 正午儒略日數。

`'Chinese ymd'`: 包含中曆的年月日，閏月的月份用負數表示。 

`'cyear'`: 與中曆年日期重疊最多的公曆年份，也是`'Chinese ymd'` list 中第一個數字。此例的太平真君六年就是445年，因為太平真君六年始於445年1月24日，終於446年2月11日，太平真君六年與公曆445年有342日重疊，是重疊日數最多的公曆年，因此甲辰年的`'cyear'`取445。

`'cdate'`: 中曆日期。

`'sexagenary year'`: python list 顯示年干支，年地支對應的生肖，及干支數 (1=甲子，2＝乙丑，3＝丙寅等，六十干支的干支數可參閱[干支網頁](https://ytliu0.github.io/ChineseCalendar/sexagenary_chinese.html)的表二) 。

`'jian'`: 中曆月的月建干支，月干支的詳情可參閱[干支網頁](https://ytliu0.github.io/ChineseCalendar/sexagenary_chinese.html)。

`'Nday_cmonth'`: 中曆月的日數，一般是29日或30日。

`'sexagenary date'`: 日干支及其干支數，日干支的詳情參閱[干支網頁](https://ytliu0.github.io/ChineseCalendar/sexagenary_chinese.html)。

`'Week'`: 星期及星期數，星期數用 0、1、2、3、4、5、6 表示星期日、一、二、三、四、五、六。 

`'period'`: 中國歷史的時代。

`'Chinese calendar'`: 中曆的版本。

與我的[農曆網站](https://ytliu0.github.io/ChineseCalendar/index_chinese.html)一樣，公曆年 y 的年限是 -721 至 2200。當中國處於分裂時代時常有多個曆法同時使用，這裡用的指令只提供其中一個版本，其他版本的日期也可以獲取，方法是選'calendar'參數為'default'及'multiple_calendars'為True:

```python
cal.western_to_chinese_date(445, 2, 25, multiple_calendars=True)
```

此輸出顯示公曆445年2月25日在劉宋的曆法是文帝元嘉二十二年二月初三，當時宋朝行用的曆法是元嘉曆，此日期就是預設中曆輸出的日期。

指令`cal.western_to_chinese_year(445, 2, calendar='WeiZhouSui')`輸出 python list 包含公曆445年2月對應的北魏曆法日期；`cal.western_to_chinese_year(445, calendar='WeiZhouSui')` 則輸出 python list 包含公曆445年全年對應的北魏曆法日期。

```python
cal.western_to_chinese_year(445, 2, calendar='WeiZhouSui')
```

以上的輸出如果用於其後的日曆計算或許有用，但對於瀏覽來說並不是很方便，函數`.western_to_chinese_year_html()`輸出的 HTML 文字用`IPython.display.HTML()`的函數處理後，就會得到如[這農曆網頁](https://ytliu0.github.io/ChineseCalendar/index_chinese.html)的表列顯示，比較方便瀏覽。

```python
# 公曆445年2月與北魏日曆對照
HTML(cal.western_to_chinese_year_html(445, 2, calendar='WeiZhouSui'))
```

```python
# 顯示445年2月的北魏月曆，並顯示每日的正午儒略日數。
HTML(cal.western_to_chinese_year_html(445, 2, calendar='WeiZhouSui', showJDN=True))
```

上面列出的月相和二十四節氣如[農曆網站](https://ytliu0.github.io/ChineseCalendar/index_chinese.html)的一樣是指用現代方法以噴射推進實驗室的 DE441 曆表算出，而曆書節氣指按當時中國曆法算出的節氣日期。

中國科學院紫金山天文臺從 2014年起用 DE421 曆表編算天文年曆，直至2025年還沒有更新曆表，不過 DE421 算出的月相和節氣時刻與 DE431/DE441 在現代相差甚微。所有用現代曆表算出的月相和節氣時刻都用東八區時間 UT+8 (UT 在1972年前指 UT1， 1972年到今年指 UTC，未來年份則是預推的 UTC)。

```python
# 用國標 《農曆的編算和頒行》 GB/T 33661-2017 算出的2500年年曆
HTML(cal.western_to_chinese_year_html(2500, calendar='GBT'))
```

## 中曆轉公曆

本軟體提供以下四個函數可用來轉中曆為公曆：

`.chinese_to_western_date(y, m, d, calendar='default', ephemeris='None')`: 轉中曆年月日 y, m, d 為公曆日期，閏月的 m 值為負數。

`.chinese_year_info(y, calendar='default', ephemeris='None')`: 函數輸出中曆年 y 的所有中曆月初一的公曆日期，曆書節氣，及用現代曆表 (DE431 或 DE441) 算出的節氣。

`.chinese_year_html(y, calendar='default', ephemeris='None')`: 與上述函數的功能一致，但輸出 HTML 文字。

`.chinese_to_western_year(y, m=None, calendar='default', ephemeris='None')`: 列出中曆年 y 中曆月 m 所有日期的公曆日期。如果 m 是 None，則列出中曆年全年的公曆日期。

`chinese_to_western_year_html(y, m=None, calendar='default', showJDN=False, ephemeris='None', show_chinese_year_info=True, show_chinese_calendar_info=True)`: 與上述函數的功能一致，但輸出 HTML 文字。

我的[農曆網站](https://ytliu0.github.io/ChineseCalendar/index_chinese.html) 並沒有網頁列出這些函數提供的公曆資料，[朔閏表網頁](https://ytliu0.github.io/ChineseCalendar/table_chinese.html)提供與`.chinese_year_html()`函數相近的資料。

這些函數的用法與公曆轉中曆的函數接近，茲舉例展示。

```python
# 轉南明永曆十五年(1661年)八月十五日為公曆
cal.chinese_to_western_date(1661, 8, 15, calendar='SouthernMing')
```

輸出顯示公曆日期是1661年9月8日。注意南明的八月十五在這年與清朝時憲曆不一致，清世祖順治十八年八月十五在1661年10月7日，可用`cal.chinese_to_western_date(1661, 8, 15)`驗證。

歷代中國因曆法改革有時會出現中曆日期對應兩個公曆日期的情況，例如762年的五月十二日：

```python
cal.chinese_to_western_date(762, 5, 12)
```

以下指令顯示顓頊曆-250年(前251年)的曆日資料。

```python
cal.chinese_year_info(-250, calendar='Zhuanxu')
```

此函數輸出 python dictionary，有以下字鍵：

`'cm'`: python list 含有以下中曆資料： \[{'cm':m1, 'JDN':JDN1, 'jian':'jian1', 'n':n1}, {'cm':m2, 'JDN':JDN2, 'jian':'jian2', 'n':n2}, ...\]，'cm' 表示中曆月序(負數表示閏月), 'JDN' 表示朔日的正午儒略日數， 'jian' 表示用月建來算的月干支數 ('jian'='1' 表示甲子，'2' 表示乙丑，⋯⋯，'60' 表示癸亥，'NA' 表示中曆月沒有月建，一般是閏月，但-103年以前的曆法因不用無中氣置閏也不列月建，'NZ' 表示此月不含中氣，這只適用於古六曆的月份)，'n' 表示中曆月的日數。 此 python list 以 'JDN' 的數值由小至大排序。

`'calendar'`: 中曆的版本 (例如 '三統曆'， '時憲曆' 等)。

`'reign/era'` (只適用於 1912年前的年份): 中曆年的年號或皇帝紀年。

`'calst'`: 曆書節氣 tuples \[(lab1, jdn1), (lab2, jdn2), ...\]， lab1, lab2, ... 是曆書節氣的標記，例如 'J1' 表示正月節(立春)，'Z4' 表示四月中(小滿) 等，jdn1, jdn2, ... 是曆書節氣的正午儒略日數。

`'datongST'` (只適用於預設中曆在1667至1669的中曆年): 用大統曆算出的節氣 \[(lab1, jdn1), (lab2, jdn2), ...\]。康熙六年(1667)至八年(1669)清朝因曆獄而復用明朝《大統曆》，但後來清朝欽天監編的《御定萬年書》試圖抹去這段歷史，用時憲曆追推這幾年的節氣，但臺灣國立故宮博物院的清代檔案檢索系統藏有這幾年的《大清七政經緯躔度時憲曆》，顯示這幾年的節氣日期確是符合大統曆的推算，而不符合時憲曆的推算，所以此處在這幾年列出大統曆的節氣。

`'sterm'`: python list \[\[lab1, info1\], \[lab2, info2\], ...\] 顯示用 DE431/DE441曆表算出的節氣，lab1, lab2, ... 為節氣標記，info1, info2, ... 則是 python dictionary，顯示節氣的日期及 UT+8 時刻。

`'mphase'`: python list \[\[lab1, info1\], \[lab2, info2\], ...\] 顯示用 DE431/DE441 曆表算出的中曆年內所有的月相資料。lab1, lab2, ... 是月相的標記： 'Q0' 是朔，'Q1' 是上弦，'Q2' 是望，'Q3' 是下弦。 info1, info2, ... 是 python dictionary，顯示月相的日期及 UT+8 時刻。

此函數輸出的 python list 不方便瀏覽， 函數 `.chinese_year_info_html()` 輸出 HTML 文字，可用 `IPython.display.HTML()` 函數將資料以表格形式顯示。

```python
# 顓頊曆-250年(前251年)的曆日資料
HTML(cal.chinese_year_html(-250, calendar='Zhuanxu'))
```

```python
# 用國標《農曆的編算與頒行》 GB/T 33661-2017 算出的2714年中曆
HTML(cal.chinese_year_html(2714, calendar='GBT'))
```

```python
# 遼穆宗應曆五年(955年)八月的所有日期
cal.chinese_to_western_year(955, 8, calendar='LiaoJinYuan')
```

```python
# 遼穆宗應曆五年(955年)八月的 HTML 月曆表
HTML(cal.chinese_to_western_year_html(955, 8, calendar='LiaoJinYuan'))
```

`.western_to_chinese_year_html()`函數輸出的月相和節氣是公曆日期，此數輸出的則是中曆日期。

## 月相和二十四節氣

`ChineseCalendar_py.sunMoon.sunMoon` class 提供用現代曆表算出的月相和二十四節氣的 UT+8 日期和時刻，以下指令建立這 class 的實體。

```python
sm = ccal.sunMoon.sunMoon('ChiT')
```

第一個輸入參數 'ChiT' 表示輸出的語言是傳統中文，其他可選的是 'Eng'(英文)和 'ChiS' (簡體中文)。

第二個可輸入的參數是 'DE431' 或 'DE441'， 表示計算月相和節氣用的曆表，默認值是 'DE441'。如前述，紫金山天文臺自2014年起用 DE421 曆表編算天文年曆，直至2025年仍是如此，DE421、DE431及DE441曆表算出的月相和節氣在距今幾十年間差異甚微(DE421 曆表的年限是1899年至2053年)，在遠古和遙遠將來 DE431 和 DE441 算出的時刻雖然差異較大，但仍未超出 TDB 轉換 UT1 的估計誤差。1929年至2013年中曆的月相和節氣用舊曆表計算，與 DE 曆表的差異可達半分鐘。1929年前的節氣不但用舊曆表計算，時刻也不用UT+8，差異當然更大。

-3502-3503月相與節氣的TDB時刻已預先算好，數據存於源碼資料夾`src/ChineseCalendar_py/sunMoon_DE431.csv`和`src/ChineseCalendar_py/sunMoon_DE441.csv`。在源碼資料夾`src/ChineseCalendar_py/`有函數`get_dTT_UT()`輸出 TT-UT 的數值，用於轉月相和節氣時刻為UT+8，此函數大約一年要更新一次以含括最新閏秒數據，因閏秒將於2035年或更早會被取消，到時便不用頻繁更新。

```python
# 獲取公元3500年的節氣和月相
st, mp = sm.st_mp(3500)
print('節氣:\n', st)
print('月相:\n', mp)
```

```python
# 公元3500年的全年的月相和節氣，轉為 pandas dataframe 來顯示。
st, mp = sm.st_mp_df(3500)
display(st)
display(mp)
```

月相中朔和望還提供日月食的資料，有連結到我製的日月食網頁，要恰當顯示連結，可用 pandas dataframe 提供的 `.to_html` 函數轉為 HTML 文字，然後用 `IPython.display.HTML()` 顯示，得出的輸出類此[氣朔網頁](https://ytliu0.github.io/ChineseCalendar/sunMoon_chinese.html)的表格。

```python
HTML(mp.to_html(
    render_links=True,
    escape=False,
))
```

## 曆書節氣

```python
# 吳大帝赤烏十三年(250年)的曆書節氣
cal.calendrical_solar_terms(250, calendar='Wu')
```

```python
# 吳大帝赤烏十三年(250年)的曆書節氣轉為 pandas dataframe 顯示
display(pd.DataFrame(cal.calendrical_solar_terms(250, calendar='Wu')))
```

此函數輸出公曆 y 年的曆書節氣，如要中曆 y 年的曆書節氣，可用上述
`.chinese_year_info()` 或 `.chinese_year_html()` 函數。
