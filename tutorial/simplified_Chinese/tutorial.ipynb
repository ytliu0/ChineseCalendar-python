## 预设中历及其他支持的中历

中国有时处于分裂局面，各政权行用不同的历法，本软体在分裂时代提供多部（但非全部）同时行用的历法。

为免混乱，如果用者没有指明用某一特定历法，本软体在中国分裂时代选取了一套历法，称为「预设中历」。预设中历在春秋时代用鲁国历法（春秋历），战国时代用周朝历法，三国时代用魏国历法，东晋南北朝用南朝（东晋、宋、齐、梁、陈）历法，五代十国用北朝（梁、唐、晋、汉、周）历法，宋、辽、金、蒙古时期用宋朝历法，南明、明郑、清时期用清朝历法，详情可参阅[预设中历网页](https://ytliu0.github.io/ChineseCalendar-python/tutorial/simplified_Chinese/default_calendar.html)，预设中历的年限是-721年至2200年。

如[预设中历网页](https://ytliu0.github.io/ChineseCalendar-python/tutorial/simplified_Chinese/default_calendar.html)所述，-103年至2200年的中历用查表法获取，这段时期中历的朔闰及历书节气数据存于源码文件夹'src/ChineseCalendar_py/'的'default_calendar_M105_2201.csv'档案中，历日以距公历年1月0日的日数表示。当中历物件实例建立时，档案中的朔闰数据载入 numpy array `.calData`，历书节气数据载入 numpy array `.calSterm`。在这资料夹的日历计算习题 notebook 中会有实例展示这两个 numpy array 在某些计算上很有用。为了方便计算日历，这两个 numpy array 的历日资料其实涵盖-105年至2202年。

除了预设中历外，本软体还提供一些其他的中历，详情可参阅[其他中历网页](https://ytliu0.github.io/ChineseCalendar-python/tutorial/simplified_Chinese/available_calendars.html)。网页中列出的所有中历都有行用年份，大多数中历用其历法法则计算，原则上可用于行用年份之外，但这里极不鼓励用者如此做。

## 公历转中历

本软体套件提供下列三个函数转公历为中历：

- `.western_to_chinese_date(y, m, d, calendar='default', multiple_calendars=False, ephemeris='None')`: 公历日期(年 = y, 月 = m, 日 = d)转中历日期。

- `.western_to_chinese_year(y, m=None, calendar='default', ephemeris='None')`: 公历 y 年 m 月 中所有日期转为中历日期。如果 m 是 None，则将公历年 y 中所有日期转为中历日期。

- `.western_to_chinese_year_html(y, m=None, calendar='default', showJDN=False, ephemeris='None', show_chinese_year_info=True,show_chinese_calendar_info=True)`: 与函数`.western_to_chinese_year(y, m=None, calendar='default', ephemeris='None')`的功能一样，但会输出 HTML 格式的文字。

`calendar`参数用来选取某中历，默认值是'default'，表示预设中历，用者可选择[其他中历网页](https://ytliu0.github.io/ChineseCalendar-python/tutorial/simplified_Chinese/available_calendars.html)列出的中历标记选择其他中历。

`multiple_calendars`参数只有当`calendar`参数是'default'(预设中历)时才有作用。如果`multiple_calendars`设定为`True`，分裂时代的输出中历会列出多个当时行用的中历。

`ephemeris`参数让用者选择用哪个现代历表计算月相和节气时刻，这里只支持'DE431'和'DE441'，默认值是'None'，表示会用已设定的历表，如果还未设定历表，则用默认值'DE441'。

以下指令转公历445年2月25日为北魏行用的历法。根据[其他中历网页](https://ytliu0.github.io/ChineseCalendar-python/tutorial/simplified_Chinese/available_calendars.html)，北魏的日历要用 'WeiZhouSui' 标记。

```python
cal.western_to_chinese_date(445, 2, 25, calendar='WeiZhouSui')
```

函数输出显示公历445年2月25日，在北魏的日历是太武帝太平真君六年闰正月初三，年干支是乙酉、日干支是癸亥，当时北魏行用的历法是景初历。

此函数的输出是 python dictionary，有下列的字典键(keys):

`'Western ymd'`: python list 包含公历的年月日。

`'Western calendar'`: 公历版本 (格里历、儒略历或逆推儒略历)。

`'JDN'`: 正午儒略日数。

`'Chinese ymd'`: 包含中历的年月日，闰月的月份用负数表示。 

`'cyear'`: 与中历年日期重叠最多的公历年份，也是`'Chinese ymd'` list 中第一个数字。此例的太平真君六年就是445年，因为太平真君六年始于445年1月24日，终于446年2月11日，太平真君六年与公历445年有342日重叠，是重叠日数最多的公历年，因此甲辰年的`'cyear'`取445。

`'cdate'`: 中历日期。

`'sexagenary year'`: python list 显示年干支，年地支对应的生肖，及干支数 (1=甲子，2＝乙丑，3＝丙寅等，六十干支的干支数可参阅[干支网页](https://ytliu0.github.io/ChineseCalendar/sexagenary_simp.html)的表二) 。

`'jian'`: 中历月的月建干支，月干支的详情可参阅[干支网页](https://ytliu0.github.io/ChineseCalendar/sexagenary_simp.html)。

`'Nday_cmonth'`: 中历月的日数，一般是29日或30日。

`'sexagenary date'`: 日干支及其干支数，日干支的详情参阅[干支网页](https://ytliu0.github.io/ChineseCalendar/sexagenary_simp.html)。

`'Week'`: 星期及星期数，星期数用 0、1、2、3、4、5、6 表示星期日、一、二、三、四、五、六。 

`'period'`: 中国历史的时代。

`'Chinese calendar'`: 中历的版本。

与我的[农历网站](https://ytliu0.github.io/ChineseCalendar/index_simp.html)一样，公历年 y 的年限是 -721 至 2200。当中国处于分裂时代时常有多个历法同时使用，这里用的指令只提供其中一个版本，其他版本的日期也可以获取，方法是选'calendar'参数为'default'及'multiple_calendars'为True:

```python
cal.western_to_chinese_date(445, 2, 25, multiple_calendars=True)
```

此输出显示公历445年2月25日在刘宋的历法是文帝元嘉二十二年二月初三，当时宋朝行用的历法是元嘉历，此日期就是预设中历输出的日期。

指令`cal.western_to_chinese_year(445, 2, calendar='WeiZhouSui')`输出 python list 包含公历445年2月对应的北魏历法日期；`cal.western_to_chinese_year(445, calendar='WeiZhouSui')` 则输出 python list 包含公历445年全年对应的北魏历法日期。

```python
cal.western_to_chinese_year(445, 2, calendar='WeiZhouSui')
```

以上的输出如果用于其后的日历计算或许有用，但对于浏览来说并不是很方便，函数`.western_to_chinese_year_html()`输出的 HTML 文字用`IPython.display.HTML()`的函数处理后，就会得到如[这农历网页](https://ytliu0.github.io/ChineseCalendar/index_simp.html)的表列显示，比较方便浏览。

```python
# 公历445年2月与北魏日历对照
HTML(cal.western_to_chinese_year_html(445, 2, calendar='WeiZhouSui'))
```

```python
# 显示445年2月的北魏月历，并显示每日的正午儒略日数。
HTML(cal.western_to_chinese_year_html(445, 2, calendar='WeiZhouSui', showJDN=True))
```

上面列出的月相和二十四节气如[农历网站](https://ytliu0.github.io/ChineseCalendar/index_simp.html)的一样是指用现代方法以喷射推进实验室的 DE441 历表算出，而历书节气指按当时中国历法算出的节气日期。

中国科学院紫金山天文台从 2014年起用 DE421 历表编算天文年历，直至2025年还没有更新历表，不过 DE421 算出的月相和节气时刻与 DE431/DE441 在现代相差甚微。所有用现代历表算出的月相和节气时刻都用东八区时间 UT+8 (UT 在1972年前指 UT1， 1972年到今年指 UTC，未来年份则是预推的 UTC)。

```python
# 用国标 《农历的编算和颁行》 GB/T 33661-2017 算出的2500年年历
HTML(cal.western_to_chinese_year_html(2500, calendar='GBT'))
```

## 中历转公历

本软体提供以下四个函数可用来转中历为公历：

`.chinese_to_western_date(y, m, d, calendar='default', ephemeris='None')`: 转中历年月日 y, m, d 为公历日期，闰月的 m 值为负数。

`.chinese_year_info(y, calendar='default', ephemeris='None')`: 函数输出中历年 y 的所有中历月初一的公历日期，历书节气，及用现代历表 (DE431 或 DE441) 算出的节气。

`.chinese_year_html(y, calendar='default', ephemeris='None')`: 与上述函数的功能一致，但输出 HTML 文字。

`.chinese_to_western_year(y, m=None, calendar='default', ephemeris='None')`: 列出中历年 y 中历月 m 所有日期的公历日期。如果 m 是 None，则列出中历年全年的公历日期。

`chinese_to_western_year_html(y, m=None, calendar='default', showJDN=False, ephemeris='None', show_chinese_year_info=True, show_chinese_calendar_info=True)`: 与上述函数的功能一致，但输出 HTML 文字。

我的[农历网站](https://ytliu0.github.io/ChineseCalendar/index_simp.html) 并没有网页列出这些函数提供的公历资料，[朔闰表网页](https://ytliu0.github.io/ChineseCalendar/table_simp.html)提供与`.chinese_year_html()`函数相近的资料。

这些函数的用法与公历转中历的函数接近，兹举例展示。

```python
# 转南明永历十五年(1661年)八月十五日为公历
cal.chinese_to_western_date(1661, 8, 15, calendar='SouthernMing')
```

输出显示公历日期是1661年9月8日。注意南明的八月十五在这年与清朝时宪历不一致，清世祖顺治十八年八月十五在1661年10月7日，可用`cal.chinese_to_western_date(1661, 8, 15)`验证。

历代中国因历法改革有时会出现中历日期对应两个公历日期的情况，例如762年的五月十二日：

```python
cal.chinese_to_western_date(762, 5, 12)
```

以下指令显示颛顼历-250年(前251年)的历日资料。

```python
cal.chinese_year_info(-250, calendar='Zhuanxu')
```

此函数输出 python dictionary，有以下字键：

`'cm'`: python list 含有以下中历资料： \[{'cm':m1, 'JDN':JDN1, 'jian':'jian1', 'n':n1}, {'cm':m2, 'JDN':JDN2, 'jian':'jian2', 'n':n2}, ...\]，'cm' 表示中历月序(负数表示闰月), 'JDN' 表示朔日的正午儒略日数， 'jian' 表示用月建来算的月干支数 ('jian'='1' 表示甲子，'2' 表示乙丑，⋯⋯，'60' 表示癸亥，'NA' 表示中历月没有月建，一般是闰月，但-103年以前的历法因不用无中气置闰也不列月建，'NZ' 表示此月不含中气，这只适用于古六历的月份)，'n' 表示中历月的日数。 此 python list 以 'JDN' 的数值由小至大排序。

`'calendar'`: 中历的版本 (例如 '三统历'， '时宪历' 等)。

`'reign/era'` (只适用于 1912年前的年份): 中历年的年号或皇帝纪年。

`'calst'`: 历书节气 tuples \[(lab1, jdn1), (lab2, jdn2), ...\]， lab1, lab2, ... 是历书节气的标记，例如 'J1' 表示正月节(立春)，'Z4' 表示四月中(小满) 等，jdn1, jdn2, ... 是历书节气的正午儒略日数。

`'datongST'` (只适用于预设中历在1667至1669的中历年): 用大统历算出的节气 \[(lab1, jdn1), (lab2, jdn2), ...\]。康熙六年(1667)至八年(1669)清朝因历狱而复用明朝《大统历》，但后来清朝钦天监编的《御定万年书》试图抹去这段历史，用时宪历追推这几年的节气，但台湾国立故宫博物院的清代档案检索系统藏有这几年的《大清七政经纬躔度时宪历》，显示这几年的节气日期确是符合大统历的推算，而不符合时宪历的推算，所以此处在这几年列出大统历的节气。

`'sterm'`: python list \[\[lab1, info1\], \[lab2, info2\], ...\] 显示用 DE431/DE441历表算出的节气，lab1, lab2, ... 为节气标记，info1, info2, ... 则是 python dictionary，显示节气的日期及 UT+8 时刻。

`'mphase'`: python list \[\[lab1, info1\], \[lab2, info2\], ...\] 显示用 DE431/DE441 历表算出的中历年内所有的月相资料。lab1, lab2, ... 是月相的标记： 'Q0' 是朔，'Q1' 是上弦，'Q2' 是望，'Q3' 是下弦。 info1, info2, ... 是 python dictionary，显示月相的日期及 UT+8 时刻。

此函数输出的 python list 不方便浏览， 函数 `.chinese_year_info_html()` 输出 HTML 文字，可用 `IPython.display.HTML()` 函数将资料以表格形式显示。

```python
# 颛顼历-250年(前251年)的历日资料
HTML(cal.chinese_year_html(-250, calendar='Zhuanxu'))
```

```python
# 用国标《农历的编算与颁行》 GB/T 33661-2017 算出的2714年中历
HTML(cal.chinese_year_html(2714, calendar='GBT'))
```

```python
# 辽穆宗应历五年(955年)八月的所有日期
cal.chinese_to_western_year(955, 8, calendar='LiaoJinYuan')
```

```python
# 辽穆宗应历五年(955年)八月的 HTML 月历表
HTML(cal.chinese_to_western_year_html(955, 8, calendar='LiaoJinYuan'))
```

`.western_to_chinese_year_html()`函数输出的月相和节气是公历日期，此数输出的则是中历日期。

```python
# 蜀后主延熙十七年(254年)的年历表
HTML(cal.chinese_to_western_year_html(254, calendar='Shu'))
```

## 月相和二十四节气

`ChineseCalendar_py.sunMoon.sunMoon` class 提供用现代历表算出的月相和二十四节气的 UT+8 日期和时刻，以下指令建立这 class 的实体。

```python
sm = ccal.sunMoon.sunMoon('ChiS')
```

第一个输入参数 'ChiS' 表示输出的语言是简体中文，其他可选的是 'Eng'(英文)和 'ChiT' (传统中文)。

第二个可输入的参数是 'DE431' 或 'DE441'， 表示计算月相和节气用的历表，默认值是 'DE441'。如前述，紫金山天文台自2014年起用 DE421 历表编算天文年历，直至2025年仍是如此，DE421、DE431及DE441历表算出的月相和节气在距今几十年间差异甚微(DE421 历表的年限是1899年至2053年)，在远古和遥远将来 DE431 和 DE441 算出的时刻虽然差异较大，但仍未超出 TDB 转换 UT1 的估计误差。1929年至2013年中历的月相和节气用旧历表计算，与 DE 历表的差异可达半分钟。1929年前的节气不但用旧历表计算，时刻也不用UT+8，差异当然更大。

-3502-3503月相与节气的TDB时刻已预先算好，数据存于源码资料夹`src/ChineseCalendar_py/sunMoon_DE431.csv`和`src/ChineseCalendar_py/sunMoon_DE441.csv`。在源码资料夹`src/ChineseCalendar_py/`有函数`get_dTT_UT()`输出 TT-UT 的数值，用于转月相和节气时刻为UT+8，此函数大约一年要更新一次以含括最新闰秒数据，因闰秒将于2035年或更早会被取消，到时便不用频繁更新。

```python
# 获取公元3500年的节气和月相
st, mp = sm.st_mp(3500)
print('节气:\n', st)
print('月相:\n', mp)
```

```python
# 公元3500年的全年的月相和节气，转为 pandas dataframe 来显示。
st, mp = sm.st_mp_df(3500)
display(st)
display(mp)
```

月相中朔和望还提供日月食的资料，有连结到我制的日月食网页，要恰当显示连结，可用 pandas dataframe 提供的 `.to_html` 函数转为 HTML 文字，然后用 `IPython.display.HTML()` 显示，得出的输出类此[气朔网页](https://ytliu0.github.io/ChineseCalendar/sunMoon_simp.html)的表格。

```python
HTML(mp.to_html(
    render_links=True,
    escape=False,
))
```

## 历书节气

古代的节气用当时的历法计算，算出的节气与用现代历表的节气常有大差异，函数`.calendrical_solar_terms(y, calendar='default', ephemeris='None')`提供公历 y 年的历书节气，即按当时历法计算的节气。参数`ephemeris`只有当选取 calendar='GBT' 时用来选择DE历表，其他的calendar选项没有作用。

```python
# 吴大帝赤乌十三年(250年)的历书节气
cal.calendrical_solar_terms(250, calendar='Wu')
```

```python
# 吴大帝赤乌十三年(250年)的历书节气转为 pandas dataframe 显示
display(pd.DataFrame(cal.calendrical_solar_terms(250, calendar='Wu')))
```

此函数输出公历 y 年的历书节气，如要中历 y 年的历书节气，可用上述
`.chinese_year_info()` 或 `.chinese_year_html()` 函数。
